```python
class CustomTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False):
        # Pop labels from inputs and convert them to the format expected by the loss function
        labels = inputs.pop("labels").long()

        # Call model forward method
        outputs = model(**inputs)

        # Compute the loss
        logits = outputs.logits
        # Shift the logits and labels for the loss calculation
        shift_logits = logits[..., :-1, :].contiguous()
        shift_labels = labels[..., 1:].contiguous()

        # Flatten the logits and labels to compute loss
        loss_fct = torch.nn.CrossEntropyLoss()
        loss = loss_fct(shift_logits.view(-1, shift_logits.size(-1)), shift_labels.view(-1))

        return (loss, outputs) if return_outputs else loss
```
